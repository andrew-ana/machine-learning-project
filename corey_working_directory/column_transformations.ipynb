#### Convert 'ccyymmdd' to abbreviation for month name

```python
airline[8] = airline[8].astype(str)
airline_list = airline[8].to_list()
month = [i[4:6] for i in airline_list]
month_dict = {'01':'jan', '02':'feb', '03':'mar', '04':'apr', '05':'may', '06':'jun', '07':'jul', '08':'aug', '09':'sep', '10':'oct', '11':'nov', '12':'dec'}
month_dict2 = [month_dict[j] for j in month]
airline[8] = month_dict2  # or add to end of df
```

####  Change numeric day of week to abbreviated name

```python
day_dict = {1:'mon', 2:'tue', 3:'wed', 4:'thu', 5:'fri', 6:'sat', 7:'sun'}
day_list = airline[9].to_list()
day_name = [day_dict[day_num] for day_num in day_list]
airline[9] = day_name  # or add to end of df
```

#### Make factor-valued flag indicating whether a flight was late

```python
arrival_delay = airline[21].to_list()

arrival_flag = []

for i in arrival_delay:
    if i > 0:
        arrival_flag.append(1)
    else:
        arrival_flag.append(0)

airline[21] = arrival_flag  # or add to end of df
```

#### Change departure time to time blocks

```python
departure_time_list = airline[10].to_list()

departure_block = []

for i in departure_time_list:
    if 0 <= i <= 259:
        departure_block.append("late_night")
    elif 300 <= i <= 559:
        departure_block.append("early_morning")
    elif 600 <= i <= 859:
        departure_block.append("morning")
    elif 900 <= i <= 1159:
        departure_block.append("late_morning")
    elif 1200 <= i <= 1459:
        departure_block.append("early_afternoon")
    elif 1500 <= i <= 1759:
        departure_block.append("late_afternoon")
    elif 1800 <= i <= 2059:
        departure_block.append("evening")
    elif 2100 <= i <= 2359:
        departure_block.append("late_evening")

airline[10] = departure_block # or add new column
```
